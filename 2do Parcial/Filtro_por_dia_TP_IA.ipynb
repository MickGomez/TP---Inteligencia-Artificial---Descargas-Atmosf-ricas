```python
import os
import pandas as pd

# Define el directorio principal en Google Drive
base_dir = '/content/drive/MyDrive/2. TP Inteligencia Artificial/DMH/2016'

# Lista para almacenar los DataFrames de cada mes
data_frames = []

# Recorre cada subcarpeta (mes) en el directorio principal
for month_dir in os.listdir(base_dir):
    month_path = os.path.join(base_dir, month_dir)
    if os.path.isdir(month_path):
        print(f"Procesando directorio: {month_path}")
        # Recorre cada archivo en la subcarpeta
        for file_name in os.listdir(month_path):
            file_path = os.path.join(month_path, file_name)
            if file_path.endswith('.csv'):
                print(f"Leyendo archivo: {file_path}")
                try:
                    # Lee el archivo CSV en un DataFrame
                    df = pd.read_csv(file_path)
                    if not df.empty:
                        # Asegúrate de que la columna de fecha esté en formato datetime
                        df['timestamp'] = pd.to_datetime(df['timestamp'])
                        # Añade el DataFrame a la lista
                        data_frames.append(df)
                    else:
                        print(f"Archivo vacío: {file_path}")
                except Exception as e:
                    print(f"Error al leer el archivo {file_path}: {e}")

# Verifica si hay DataFrames para combinar
if data_frames:
    # Combina todos los DataFrames en uno solo
    combined_df = pd.concat(data_frames)
    # Realiza el conteo de los datos por día
    count_by_day = combined_df.groupby(combined_df['timestamp'].dt.date).size().reset_index(name='count')
    # Guarda el resultado en un archivo CSV en Google Drive
    output_path = '/content/drive/My Drive/conteo_2016.csv'
    count_by_day.to_csv(output_path, index=False)
    print(f"El archivo '{output_path}' ha sido creado.")
else:
    print("No se encontraron archivos CSV para procesar.")
```
